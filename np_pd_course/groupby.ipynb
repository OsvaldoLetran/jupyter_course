### Groupby
Permite agrupar datos en función de los demás. Es decir, hacer el análisis del DataFrame en función de una de las columnas.

```python
df = pd.read_csv('/home/jaime/jupyter_course/np_pd_course/bestsellers_with_categories.csv', sep=',', header=0)
df.head(2)
```

```python
df.groupby('Author').count()    # cuantas coincidencias?
# df.groupby('Author').mean()    # media de rating
# df.groupby('Author').median()    # mediana del rating
# df.groupby('Author').min()    # minimo de rating
# df.groupby('Author').max()    # maximo de rating
# df.groupby('Author').sum()    # suma del rating
```

Ahora 'Author' es un indice de nuestro df, podemos usar algo como por ejemplo:

```python
df.groupby('Author').sum().loc['William Davis']
```

```python
df.groupby('Author').sum().reset_index()   # vuelve 'Author' parte de las columnas
```

```python
df.groupby('Author').agg('min', 'max')   # funciones de agregacion
```

```python
df.groupby('Author').agg({'Reviews':['min', 'max'], 'User Rating':'sum'})
```

```python
df.groupby(['Author', 'Year']).count()   # nuestro index pasa a ser 'Author' y 'Year'
```
