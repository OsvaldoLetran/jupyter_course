## Merge

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
    'A' : ['A0', 'A1', 'A2','A3'],
    'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
    'C' : ['C0', 'C1', 'C2','C3'],
    'D': ['D0', 'D1', 'D2','D3']})
```

Unir dataframe der a izq

```python
izq.merge(der, on='key')
```

```python
# MERGE 2
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
    'A' : ['A0', 'A1', 'A2','A3'],
    'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
    'C' : ['C0', 'C1', 'C2','C3'],
    'D': ['D0', 'D1', 'D2','D3']})
```

Hay diferencias entre algunas columnas, por esa razón hay que separarlos de esta manera

```python
izq.merge(der, left_on = 'key', right_on='key_2')    # cuando las columnas tienen nombres distintos
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
    'A' : ['A0', 'A1', 'A2','A3'],
    'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2', np.nan],
    'C' : ['C0', 'C1', 'C2','C3'],
    'D': ['D0', 'D1', 'D2','D3']})
```

Si tenemos un NaNen nuestro DataFrame, pandas no lo detectará como un mach. Se soluciona con How, dando así, una preferencia.

```python
izq.merge(der, left_on = 'key', right_on='key_2', how='left')    # how='inner' viene por defecto.
```

### Join 
Es otra herramienta para hacer exactamente lo mismo, una combinación. La diferencia es que join va a ir a los índices y no a columnas específicas. Index Match

```python
izq = pd.DataFrame({'A': ['A0','A1','A2'],
  'B':['B0','B1','B2']},
  index=['k0','k1','k2'])

der =pd.DataFrame({'C': ['C0','C1','C2'],
  'D':['D0','D1','D2']},
  index=['k0','k2','k3'])
```

Combinamos izq con der

```python
izq.join(der)
```

```python
izq.join(der, how = 'inner')
```

Traer todos los datos aunque no hagan match.

```python
izq.join(der, how = 'outer')
```
