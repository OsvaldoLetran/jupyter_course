```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Con Joinplot
Joinplot une 2 distintas gr√°ficas en una sola

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue = 'sex')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue = 'sex', kind='hist')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue = 'sex', kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue='sex', kind= 'hist', 
                marginal_ticks=True, #muestra una peque√±a tabla para la gr√°fica exterior
                marginal_kws=dict(bins= 25, fill = True, multiple= 'dodge')) 
                #marginal_kws se le pueden agregar diferentes par√°metros
plt.show()
```

### Con Pairplot
Pairplot `correlaciona` todas las variables num√©ricas que hay en el dataset

```python
sns.pairplot(data= tips)
plt.show()
```

```python
sns.pairplot(data= tips, hue='sex')
plt.show()
```

`Solo muestra una esquina del pairplot, eliminando as√≠, las gr√°ficas repetidas`

```python
sns.pairplot(data= tips, hue= 'sex', corner=True)
plt.show()
```
