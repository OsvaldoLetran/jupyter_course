```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

### Uso de .kde y .displot para graficar distribuciones
Gráfico para distribuciones y que se acumulen `cumulative`

```python
sns.histplot(data = tips, x= 'tip', bins = 15, cumulative=True)
# bins=15 y cumulative=False vienen por default
plt.show()
```

Como muestra en el eje `y` `stat`, presentar los datos con `multiple`

```python
sns.histplot(data=tips, x='tip', bins=15, hue='sex', stat ='density', multiple='dodge')  # stat='count' viene por default
plt.show()
```

### Uso de .kde
Gráfica de densidad

```python
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
```

Dsitribuciones

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

### Uso de .displot
Gráfica de distribución

Cambiar tipo de gráfica con `kind`.

```python
sns.displot(data = tips, x = 'tip', hue = 'sex', kind= 'hist', multiple= 'stack')
plt.show()
```
