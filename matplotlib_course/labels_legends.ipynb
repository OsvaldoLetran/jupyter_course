```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (10,5))
ax1.plot(x,y, 'b', label = "$sin(x)$")     # con signo de dolar para que sea con notacion matematica
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
# ax1.legend(loc = "upper center")   # otras ubicaciones: upper right, lower left, right

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
plt.show()
```
