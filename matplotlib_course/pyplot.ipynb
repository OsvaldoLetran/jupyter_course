```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.linspace(0, 5, 11)
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'y')
plt.show()
```

```python
plt.plot(x, y, 'bs-')    # r de rojo, s de squad y guion para las lineas
plt.show()
```

```python
plt.plot(x, y, 'rD:')    # r de rojo, D de diamante y : para unir los diamantes ocn puntos
plt.show()
```

```python
plt.hist(x)    # histograma
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
