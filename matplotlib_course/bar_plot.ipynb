```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", "GERMANY"]
population = [1000, 800, 900, 700, 300]
```

```python
plt.bar(countrys,population, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.xticks(np.arange(5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

```python
plt.barh(countrys,population)
plt.show()
```
