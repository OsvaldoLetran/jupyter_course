```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

### Grafica de cantidad

```python
sns.countplot(data = tips, x= 'day', hue='sex')   # hue : segmentaci√≥n
plt.show()
```

```python
# el parametro dodge se utiliza en los gr√°ficos de barras para separar las barras que se superponen.
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

### Diagrama de puntos

```python
sns.swarmplot(data= tips, x= 'day', y ='total_bill', hue= 'sex', dodge=True)
plt.show()
```

### Diagrama de caja

```python
sns.boxplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True)
plt.show()
```

```python
sns.boxplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True)
sns.swarmplot(data= tips, x= 'day', y ='total_bill', hue= 'sex', dodge=True, color ='0', marker= '<')
plt.show()
```

### Viol√≠n 
parecido a Boxplot, pero no muestra los cuartiles, sino como se concentran los datos con comparaci√≥n `split`

```python
sns.violinplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True, pallette='pastel')
plt.show()
```

```python
sns.violinplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', split=True, pallette='pastel')
plt.show()
```

### Catplot
Catplot, con kind podemos cambiar el tipo de gr√°fico y se puede agregar otra variable col

```python
sns.catplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex')
plt.show()
```

```python
sns.catplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True, kind= 'box', col= 'time')
plt.show()
```
