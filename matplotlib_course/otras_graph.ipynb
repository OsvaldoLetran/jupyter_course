## Crear otros tipos de gráficos en Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

### Histograma
muestra distribución y frecuencia

```python
plt.hist(data, bins=50)   # bins=10 es por default
plt.show()
```

```python
# plt.hist(data, widht=10)    trabaja mejor default que configurando widht
plt.hist(data)
plt.show()
```

```python
plt.hist(data, histtype='step')    # histtype='bar' viene default
plt.show()
```

### Boxplot
Muestra rango intercuartil y la mediana

```python
plt.boxplot(data)
plt.show()
```

Cambiar dirección, rellenar el rango inter cuartil `patch_artist` y enfocar la mediana `notch`

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

`Outliers`

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)   # con showfliers omitimos los Outliers
plt.show()
```

### Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area, c=colors)     # s=area modifica el area de las burbujas
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors, marker='v')
plt.show()
```
