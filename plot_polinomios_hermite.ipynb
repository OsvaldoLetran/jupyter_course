```python
from scipy import special
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
def y(n):
    p_monic = special.hermite(n, monic = False)
    x = np.linspace(-2, 2, 200)
    y = p_monic(x)
    return plt.plot(x,y)
```

```python
%matplotlib notebook
y(2)  #Azul
y(3)  #Naranja
y(4)  #Verde
y(5)  #rojo
```
